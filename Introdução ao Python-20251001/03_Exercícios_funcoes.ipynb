##Distncia na terra

Construa um algoritmo que permita calcular a distncia entre dois pontos no planeta terra.
O utilizador deve introduzir a longitude (coordenada Y) e a latitude (coordenada X)  dos dois pontos em graus ().
O programa deve calcular e imprimir a distncia entre eles em Kilmetros (km).
Utilize a formula de Haversine.

https://en.wikipedia.org/wiki/Haversine_formula



```
Distncia entre dois pontos na terra

Ponto A :
Latitude 	: 39.47293741622354
Longitude 	: -8.537510346796125


Ponto B :
Latitude 	: 39.604069666367366
Longitude 	: -8.405429882228646


Distncia 	= 18.463421728067875 km
```





image.png

````python
import math

print("Distancia entre dois pontos na terra")
print("Ponto A ")
ax = float( input("Latitude  : "))
ay = float( input("Longitude : "))
print("Ponto B ")
bx = float( input("Latitude  : "))
by = float( input("Longitude : "))
#converso para radianos

x1 = math.radians(ax)
y1 = math.radians(ay)

x2 = math.radians(bx)
y2 = math.radians(by)

d = 2 * 6371*math.asin(  math.sqrt( math.sin((x2-x1)/2)**2)
+ math.cos(x1)*math.cos(x2) * math.sin((y2-y1)/2)**2)

print("Distnia = " , d)
````
