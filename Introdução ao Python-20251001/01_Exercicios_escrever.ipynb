##Olá Mundo - Várias instruções

Construa um algoritmo que imprima na consola a mensagem `"Olá Mundo"` usando várias instruções de escrever.

*NOTA:*

##Mudar de linha \n

Construa um algoritmo que imprima na consola a mensagem `"Olá Mundo dos Algoritmos"` com uma palavra em cada linha.

**NOTA:**

para introduzir caracteres especiais dentro do texto utiliza-se o caracter `\`

`\` significa que o caracter seguinte é especial.

```
\n = caracter de mudança de linha (enter)
```


````python
print( "Olá \n Mundo \n dos \n Algoritmos" )
````

````python
#por defeito o print muda de linha
print( "Olá")
print( "Mundo")
print( "dos\nAlgoritmos" , end="")
````

##Tabulação \t
Construa um algoritmo que imprima na consola o texto da figura.
```
Nº      Nome
1       Um
10      Dez
100 	Cem
1000    Mil
1000000 	 Milhão
```

**NOTA**

O caracter `\t` faz o alinhamento a cada 8 caracteres.

`|-------|-------|-------|-------|`
```
\t" = caracter tabulação
```

````python
print( "Nº \t Nome")
print( "1 \t Um")
print( "10 \t Dez")
print( "100 \t Cem")
print( "1000 \t Mil")
print( "1000000 \t Milhão")
````
