#Iteração

o ciclo **for** serve para iterar um sequencia de valores
<br>A sequência pode ser gerada: exemplo **range**
<br>ou um conjunto de dados : exemplo **Texto** ou **listas**



##iteração de um conjunto



```
for var in conjunto :
    processar var
```



````
texto = "programação em python"
print("Letras")
for letra in texto:
    print(letra, end = " ")
````

##range
<br> range - gera um sequencia de valores **inteiro**
- inicio - valor inicial (incluido)
- fim - valor final (não incluido)
- passo - distância entre os elementos
<pre>
for var in range( inicio , fim , passo) :
    processar var
</pre>




## Números entre 1 e 5 - Iteração

![flowchart0[1].png]()

````
for i in range(0, 6, 1 ):
    print(i,  " " , end="")
````

## Numeros pares

![flowchart0[1].png]()

````
#iteração de 2 em 2
for i in range(1, 10, 2 ):
    print(i,  " " , end="")
````

# Impares - Ordem inversa

![flowchart0[1].png]()

````
#iteração com valor negativo
for i in range(9, -10, - 2 ):
    print(i,  " " , end="")
````

##range - valores por defeito
`range(inicio=0, fim , passo=1)`

````
#por defeito começa em zero e o passo é 1
for i in range(5):
    print(i)
````

````
#por defeito o passo é 1
for i in range(5,10):
    print(i)
````

## Nota Mínima

![flowchart0[1].png]()

````
#importar a constante infinito e mudar-lhe o nome
from math import inf as infinito
#minimo é inicializado a infinito
minimo = infinito
#iterar seis vezes
for i in range(1, 6, 1 ):
    #solicitar a nota
    nota =  int(input( str(i) + "º nota : "))
    #verificar a a nota é menor
    if nota < minimo :
        minimo = nota;


print( "Nota mínima = ", minimo )
````
