##Cilindro
Construa um algoritmo que calcule a área da superfície e o volume de um cilindro dados o seu raio e a sua altura.

Sugestão:
- Calcule a área do círculo
- Calcule a area do retângulo





image.png

````python
import math
print("Cilindro")
h = float(input("Altura : "))
r = float(input("Raio   : "))
#calcular o volume
base = math.pi * r**2;
volume = base * h
#perimetro
perimetro = math.tau * r
area = 2*base + h * perimetro
#imprimir os resultados
print("Area =", area)
print("Volume = ", volume)

````
