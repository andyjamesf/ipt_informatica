#Somatórios
Algoritmo


```
inicializar a soma a zero
iterar a variavel do somatorio
  soma = soma + formula(variavel)
```



image.png

## Somatório

![folder[1].jpg]()![flowchart0[1].png]()

````
print( "Somatório")
soma = 0;
for i in range(1, 6, 1 ):
    print(soma,  " + ", i , end="")
    soma +=i;
    print( " \t = ", soma)
print( "\nTotal = ", soma , end="")
````

#Produtórios
Algoritmo


```
inicializar o produto a um
iterar a variavel do produtório
  produto = produto * formula(variavel)
```



# Produtório

![folder[1].jpg]()
![flowchart0[1].png]()

````
print( "Produtório")
produto = 1;
for i in range(1, 6, 1 ):
    print(produto,  " x ", i , end="")
    produto *=i;
    print( " = ", produto)
print( "\nTotal = ", produto , end="")
````
