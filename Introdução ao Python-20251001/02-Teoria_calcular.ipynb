#Operador de atribuição =
O operador de atribuição permite atribuir a uma variável um valor
- o valor pode ser o resultado de uma expressão
- o tipo da variável é deduzido a partir do tipo atribuido


````
#Atribuicao simples
a = 10
print( "a= ", a , type(a))
b = "ola"
print( "b= ", b , type(b))
c = a>5
print( "c= ", c , type(c))
````

#Atribuição múltipla
O operador de atribuição suporta múltiplas variáveis separadas por vírgulas

````
#atribuição de um valor a várias variáveis
a=b=10
print( "a= ", a , " b= ", b)

# Atribuição múltipla
c , d , e = 40, 20,30
print( "c= ", c , " d= ", d, " e = ", e)
````

# Operadores sobre inteiros


```
x + y   - soma
x - y   - subtração
x * y   - multiplicação
x / y   - divisão real
x // y  - divisão inteira
x % y   - Resto da divisão
x ** y  - potencia
```





````
 x = int(input("n1 = "))
 y = int(input("n2 = "))
 print(x ,"\t + \t" , y , "\t= " , x+y)
 print(x ,"\t - \t" , y , "\t= " , x-y)
 print(x ,"\t * \t" , y , "\t= " , x*y)
 print(x ,"\t / \t" , y , "\t= " , x/y)
 print(x ,"\t // \t" , y , "\t= " , x//y)
 print(x ,"\t % \t" , y , "\t= " , x%y)
 print(x ,"\t ** \t" , y , "\t= " , x**y)
````

# Operadores sobre Texto
```

```


````
#texto
txt1 = "|ola"
txt2 = " mundo|"
#inteiro
vezes = 5
#concatenar texto
print(txt1 + txt2)
#concatenar texto com inteiros
mensagem = txt1 + txt2 + str(vezes) + " vezes "
print(mensagem)
#replicação
print((txt1+txt2) * vezes)

````
