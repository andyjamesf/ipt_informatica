# Funções matemáticas
para utilizar as funções matemáticas é necessário fazer a sua importação para ao programa
<br>
- Importar todas as funções

```
import math
x = math.sin(3.14)
```
- Importar uma função

```
from math import sin
x = sin(3.14)
```
- Importar uma função e mudar-lhe o nome


```
from math import sin as seno
x = seno(3.14)
```




[math — Mathematical functions](https://docs.python.org/3/library/math.html)



````
import math
ang = math.pi
print( math.sin(ang))
````

````
from math import sin
print( sin(ang))
````

````
from math import sin as seno
print( seno(ang))
````

````
from math import sin as seno ,pi as PI
print( seno(PI))
````

##Arredondamento de números


```
round(x)      -> arredondamento de x
math.ceil(x)  -> arredondamento para cima de x
math.floor(x) -> arredondamento para baixo de x
```



````
valor = 12.567
print("Valor = ", valor)

#a função round faz parte da linguagem
print("Arredondamento = " , round(valor))

#importação da biblioteca matematica
import math
#funções de arredondamento da biblioteca
print("Arredonadamento para cima  = " , math.ceil(valor))
print("Arredonadamento para baixo = " , math.floor(valor))
````

## Potências


```
math.pow(base,expoente)   -> potência de base elevado a expoente
```



````
import math
base = 2
exp = 8
#operador **
print(base , "^" , exp , "=", base**exp)
#função pow
print(base , "^" , exp , "=", math.pow(base,exp))
````

## Raiz quadrada

Construa um algoritmo que calcule o valor a raiz quadrada de um número introduzido pelo utilizador

O programa deve imprimir o cálculo feito por:
- função sqrt
- função pow
- operador **

![folder[1].jpg]()

````
import math

x =  float(input("Valor de X\t: "))
print( "Raiz \t\t= ", math.sqrt( x ) )
print( "Potência \t= ", math.pow( x , 0.5 ))
print( "Operador \t= ", x ** 0.5 )
````

##Logaritmos
```
math.log10(x)       -> Logaritmo de base 10 de x
math.log(x)          -> Logaritmo natural de x (base E)
math.log(x, base)   -> Logaritmo de x na base


```

````

valor = 1.5
#logaritmo base 10
print("log10(", valor , ") = ", math.log10(valor))

#logaritmo natural
print("ln(", valor , ") = ", math.log(valor))

#logaritmo na base 5
print("log5(", valor , ") = ", math.log(valor, 5))
print("ln(", valor , ") = ", math.log(valor, math.e))
print("log10(", valor , ") = ", math.log(valor,10))
````

##Potência <=> Logaritmo

image.png

````

import math

x =  float(input("Valor de X : "))
base = float(input("Valor da Base : "))


#calcular o logaritmo
logX = math.log( x , base)
print( "Log(X) \t\t= ", logX  )

#calcular a potência
pow = base ** logX
print( "Pow(Log(X)) \t= ", pow  )
````

## Seno Coseno e Tangente



```
math.sin(ang)   -> seno do ângulo em radianos
math.cos(ang)   -> coseno do ângulo em radianos
math.tan(ang)   -> tangente do ângulo em radianos
```



folder.jpg


````
import math

ang =  float(input("Ângulo (rad) \t: "))
print( "Seno \t\t= ", math.sin( ang ))
print( "Coseno \t\t= ", math.cos( ang ))
print( "Tangente \t= ", math.tan( ang ) )
````

##conversão graus <-> radianos



```
math.radians(g) - conver o angulo g de graus para radianos
math.degrees(r) - conver o angulo r de radianos para graus
```



````
#angulo am graus
graus = 180
print("Graus = " , graus)

#converter para radianos
rad = math.radians(graus)
print("Radianos = ", rad)

#converter para graus
ang = math.degrees(rad)
print("Graus = " , ang)
````

## Trigonometria hiperbólica
```
math.sinh(ang)   -> seno hiperbólico do ângulo em radianos
math.cosh(ang)   -> coseno hiperbólico do ângulo em radianos
```

## random

##uniform
valores reais entre dois limites

```
random.uniform(a,b) -> valores reais no intervalo[a,b]
```

````
print( random.uniform(-10,10))
print( random.uniform(-10,10))
print( random.uniform(-10,10))
print( random.uniform(-10,10))
print( random.uniform(-10,10))

````

##randint
valores inteiros entre dois limites


```
random.randint(a,b) -> valor inteiro no intervalo[a,b]
```



````
print( random.randint(1,50))
print( random.randint(1,50))
print( random.randint(1,50))
print( random.randint(1,50))
print( random.randint(1,50))
print( random.randint(1,50))
````
