 * Os tipos das variáveis são deduzidos a partir dos valores que lhes são atribuídos.

```
#atribuição de um valor a uma variável
a = 10
print(a)
#type = tipo de dados de uma variavel ou valor
print(type(a))

#nova atribuicao com tipo de dados diferente
a = 12.5
print(a)
print(type(a))

#atribuição com uma expressão
a = "Hello Python " * 5
print(a)
print(type(a))

```

#Linguagem multiparadigma
A linguagem python permite programar com vários paradigmas e fazer a sua intercalação

## paradigma procedimental (procedural)

Neste paradigma o código é executado de forma sequencial.
O programa procura a primeira linha do programa ( função main, ou  primeira linha do programa) executa-a e passa para a seguinte até não haver mais linhas para executar.
Nas instruções executáveis, pode haver desvios de fluxo através de estruturas de decisão ou repetição e a execução de **funções definidas pelo utilizador**


image.png

```
#:::::::::::::::::::::::::::::::::::::::::::::::::::::::::
global saldo
#:::::::::::::::::::::::::::::::::::::::::::::::::::::::::
#depositar valores
def depositar(valor):
  global saldo
  saldo = saldo + valor
  print("Depósito ", valor ,"Saldo = " , saldo)

#Levantar valores
def levantar(valor):
  global saldo
  saldo = saldo - valor
  print("Levantamento ", valor ,"Saldo = " , saldo)
#:::::::::::::::::::::::::::::::::::::::::::::::::::::::::
if __name__ == "__main__":
    #Saldo inicial
    saldo = 100
    print("Saldo = " , saldo)
    #execução de procedimentos (funções)
    depositar(10)
    levantar(20)

```

##paradigma orientado a objetos
Um dos paradigmas mais utilizados atualmente.
O programa é executado com a criação de objetos que têm um estado próprio (atributos) que é mutável e a capacidade de executar ações (métodos). Os métodos executados interagem com outros objetos do sistema alterando os seus atributos e executando as suas ações.

[Object-Oriented Programming (OOP) in Python 3](https://realpython.com/python3-object-oriented-programming/)


image.png

```
#:::::::::::::::::::::::::::::::::::::::::::::::::::::::::
class Conta:
  #construtor
  def __init__(self,saldo):
    self.saldo = saldo
    print("Saldo = " , self.saldo)

  #depositar valores
  def depositar(self,valor):
    self.saldo = self.saldo + valor
    print("Depósito ", valor ,"Saldo = " , self.saldo)

 #Levantar valores
  def levantar(self,valor):
    self.saldo = self.saldo - valor
    print("Levantamento ", valor ,"Saldo = " , self.saldo)
#:::::::::::::::::::::::::::::::::::::::::::::::::::::::::
if __name__ == "__main__":
    #contruir um objeto
    carteira = Conta(100)
    #Executar métodos
    carteira.depositar(20)
    carteira.levantar(40)
```

## Paradigma funcional
Neste paradigma os programas são construidos aplicando e compondo funções puras, onde o seu resultado depende apenas dos seus parâmetros, sem nenhum efeito colateral observável.
Esta característica faz com que o código seja altamente paralelizavel, que permite a utilização da processamento multicore.

[Functional Programming in Python: When and How to Use It](https://realpython.com/python-functional-programming/)

```
#:::::::::::::::::::::::::::::::::::::::::::::::::::::::::
#função pura - Depositar valores
def depositar(saldo , valor):
  return saldo + valor

#função pura - Levantar valores
def levantar(saldo , valor):
  return saldo - valor

#função com argumentos que são funções
def movimento(funcao,saldo,valor):
  return funcao(saldo,valor)
#:::::::::::::::::::::::::::::::::::::::::::::::::::::::::
if __name__ == "__main__":
    #Saldo
    saldo = 100 #valor inicial
    print ("Saldo = " , saldo)

    #Depósito
    valor = 20
    saldo = depositar(saldo, valor)
    print ("Depositar ", valor , "Saldo = " , saldo)

    #Levantamento
    valor = 40
    saldo = movimento(levantar,saldo, valor)
    print ("Levantamento ", valor , "Saldo = " , saldo)
```

# The Zen of Python
O Zen of Python é uma coleção de 19 "princípios orientadores" para escrever programas de computador que influenciam o design da linguagem de programação Python.

*   Beautiful is better than ugly.
*   Explicit is better than implicit.
* Simple is better than complex.
* Complex is better than complicated.
* Flat is better than nested.
* Sparse is better than dense.
* Readability counts.
* Special cases aren't special enough to break the rules.
* Although practicality beats purity.
* Errors should never pass silently.
* Unless explicitly silenced.
* In the face of ambiguity, refuse the temptation to guess.
* There should be one-- and preferably only one --obvious way to do it.
* Although that way may not be obvious at first unless you're Dutch.
* Now is better than never.
* Although never is often better than *right* now.
* If the implementation is hard to explain, it's a bad idea.
* If the implementation is easy to explain, it may be a good idea.
* Namespaces are one honking great idea -- let's do more of those!
