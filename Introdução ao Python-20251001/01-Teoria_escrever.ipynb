# Instrução print

Documentação da função [print](https://docs.python.org/3/library/functions.html#print)
Imprime na consola
para imprimir texto os caracteres devem estar entre "texto" ou entre 'texto'


```
print("Olá mundo do python")
print('ola novamente')

```

## end

* **end** define qual a string a escrever depois de escrever os parâmetros

Por defeito o print muda de linha



```
print( "Olá mundo" , end="...")
print( " do python" , end="####")
print( " e arredores" , end=",,,,")
```

##sep
* **sep** define o caracter separador entre parâmetros


```
print(1,2,3,4,5)
print(1,2,3,4,5,sep="\t")
print(1,2,3,4,5,sep=" -> ", end = " ## ")
print(6,7,8,9,10,sep=" <- ")
```

## Caracteres especiais
- \n - mudança de linha
- \t - tabulação
- \\" - Caracter "
- \\\ - caracter \
