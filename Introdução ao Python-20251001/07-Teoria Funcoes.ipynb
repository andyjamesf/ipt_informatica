##Funções com vários parâmetros e retorno
A funções em python suportam o retorno de vários valores (separados por virgulas)

````
#funcao de calculo de vários valores
def calculo(a,b):
    soma = a+b
    dif = a-b
    #retornar dois valores
    return soma, dif

#executar a funcao
s , d = calculo( 10, 20)
print("soma = " , s , " diferença = ", d)
````

##Parâmetros nomeados
na evocação da função é possível nomear o parâmetros

````
def parametros(p1,p2,p3):
    print("parametros")
    print(" p1 = " , p1)
    print(" p2 = " , p2)
    print(" p3 = " , p3)

#execução da função
parametros(1,2,3)
#execução com nome nos parametros
parametros(p1=10,p2=20,p3=30)
#Alterar a ordem dos parametros
parametros(p3=300,p1=100,p2=200)
````

##Parâmetros por defeito
as funções podem ter parametros por defeito
- Aos parametros por defeito é atribuido um valor (=)
- os parâmetros por defeito têm de ser obrigatoriamente os últimos


```
def nomeFuncao( parametro1=valor1,parametro2=valor2,...)
```



````
def soma(n1=1 , n2=2, n3=3, n4=4):
    return n1+n2+n3+n4

print(soma(10)) #defeito n2, n3, n4
print(soma(10,20)) #defeito n3, n4
print(soma(10,20,30)) #defeito  n4
print(soma(10,20,30,40))
````

##Parâmetros com comprimento variável *
O * significa que pode ter zero ou mais parâmetros (lista de parâmetros)
* para iterar os parâmetros

```
def funcao(*parametros):
    for p in parametros:
        processar o parâmetro p
```



````
# txt - obrigatório
# parametros - zero ou mais
def mult(txt,*parametros):
    print(txt)
    m=1
    for p in parametros:
        m *= p
    return m

print(mult("sem parametros "))
print(mult("parametros 1-3",1,2,3))
print(mult("parametros 1-9",1,2,3,4,5,6,7,8,9))
````

#Documentação de funções
A documentação de uma função é feita através de docstrings

```
"""
descrição do que faz a função

"""
```



````
def juros_compostos(valor_principal, taxa, tempo):
    """
        Calcula o valor final com juros compostos.
        Os juros são incorporados no capital ao longo do tempo
        https://pt.wikipedia.org/wiki/Juro#Juros_compostos

        :param  valor_principal (float): Valor inicial investido.
        :param taxa (float): Taxa de juros anual (em decimal 10% = 0.1 ).
        :param  tempo (int): Tempo em anos.

        :return float: valor final do investimento com juros compostos.
    """
    return valor_principal * (1 + taxa) ** tempo
````

#Bibliotecas de funções
As funções devem ser guardadas em ficheiros para poderem ser reutilizadas noutros projetos

````
#execute o seguinte código para criar a minha_biblioteca
nome_biblioteca = "minha_biblioteca"
code = '''
def juros_compostos(valor_principal, taxa, tempo):
    """
        Calcula o valor final com juros compostos.
        Os juros são incorporados no capital ao longo do tempo
        https://pt.wikipedia.org/wiki/Juro#Juros_compostos

        :param  valor_principal (float): Valor inicial investido.
        :param taxa (float): Taxa de juros anual (em decimal 10% = 0.1 ).
        :param  tempo (int): Tempo em anos.

        :return float: valor final do investimento com juros compostos.
    """
    return valor_principal * (1 + taxa) ** tempo

'''
#guarda o codigo no ficheiro da biblioteca
with open(nome_biblioteca +".py","w") as file:
    file.write(code)
    print("bilioteca criada")
````

##Usar uma biblioteca do utilizador
utilizar a diretiva import

````
import minha_biblioteca
capital = 1000
juro = 0.1 #10%
anos = 10

investimento = minha_biblioteca.juros_compostos(capital,juro,anos)
print("Total do investimento = " , investimento)
````

````
from minha_biblioteca import juros_compostos
investimento = juros_compostos(capital,juro,anos)
print("Total do investimento = " , investimento)

from minha_biblioteca import juros_compostos as jc
investimento = jc(capital,juro,anos)
print("Total do investimento = " , investimento)

````

````
#importar a biblioteca
import cypherx
#encriptar um texto
texto = "Olha os romanos"
chave = 4
segredo = cypherx.Caesar(texto, chave,8).encrypt();
print("Segredo  = " , segredo)
#desencriptar o segredo
mensagem = cypherx.Caesar(segredo, chave,0).decrypt();
print("Mensagem = " , mensagem)
````
