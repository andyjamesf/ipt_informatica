# Somar intervalo - Iterativo
image.png
image.png

```
def somaI(limite):
    soma = 0;
    for i in range(1, limite + 1, 1 ):
        soma +=i;
    return soma

if __name__ == "__main__":
    lim =  int(input("Limite : "))
    print( "Soma = ", somaI(lim) )
```

# Somar intervalo - recursivo
image.pngimage.png


```
def somaR(n):
    if n > 1 :
        return n + somaR(n - 1)
    else:
        return 1


if __name__ == "__main__":
    lim =  int(input("Limite : "))
    print( "Soma = ", somaR(lim))
```

# Torres de Hanoi

image.png
image.png

```
def hanoi(orig, tmp, dest, n):
    if n > 0 :
        hanoi(orig, dest, tmp, n - 1);
        print( "\n", n,  " ", orig,  "->", dest , end="")
        hanoi(tmp, orig, dest, n - 1);


if __name__ == "__main__":
    hanoi("A", "C", "B", 3);
```

# Somar Pares - recursivo

Construa um programa que permita calcular a soma de todos os numeros PARES entre 1 e um limite introduzido pelo utilizador.
sabendo que:

- somaPar(n) = 0 se n for menor ou igual a zero
- somaPar(n) = somaPar(n-1) se n for impar
- somaPar(n) = n + somaPar(n-2) se n for par

image.png








```
def somaPar(n):
  if n < 0:
    return 0
  if n%2 != 0 :
    return somaPar(n-1)
  return n + somaPar(n-2)

somaPar(10)
```

# Sequencia de Fibonnacci - Recursivo

Construa um algoritmo que imprima os primeiros 5 numeros da sequencia de fibonacci.

image.png

```
def fibR(n):
  if n <= 2 :
    return 1
  else:
    return fibR(n-2) + fibR(n-1)

for i in range(1,6,1):
  print("fib(" , i , ") = " , fibR(i) )
```
