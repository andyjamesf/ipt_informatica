##Admitido ou Excluído
Construa um algoritmo que imprima na consola se um aluno está "Excluído" ou "Admitido" a exame dada a sua nota.

REGRA:
Um aluno está excluído se a sua nota for inferior a 7.0 e admitido caso contrário.


exemplo 1
```
Nota 	: 19.726
Admitido
```
exemplo 2


```
Nota 	: 5.177
Excluído
```






````
nota = float(input("nota : "))
if nota < 7 :
    print("Excluído")
else:
    print("Admitido")
````

##Média dos melhores dois testes

A disciplina de Algoritmos têm três testes de avaliação.<br>
A nota final é calculada pela média dos dois testes com melhor classificação.<br>
Contrua um algoritmo que solicite três a nota dos testes e de seguida imprima nota final.



```
1ª Nota 	: 24.726
2ª Nota 	: 31.207
3ª Nota 	: 3.889
Nota Final 	= 27.9665
```




````
n1 = float(input("1ª nota : "))
n2 = float(input("2ª nota : "))
n3 = float(input("3ª nota : "))

#n1 e a menor nota
if n1 <= n2 and n1 <= n3:
    #media de n2 e n3
    media = (n2 + n3) / 2
#n2 e a menor nota
elif n2 <= n1 and n2 <= n3:
    media = (n1+n3)/2
else:
    media = (n1 + n2)/2

print("media" , media)



````

## Aluno mais velho

Construa um algoritmo que solicite a data de nascimento de dois alunos.
De seguida o algoritmo deve imprimir o aluno mais velho ou a mensagem "Mesma idade".



![folder[1].jpg]()


````
print("Aluno 1")
a1 = int(input("Ano :"))
m1 = int(input("Mês :"))
d1 = int(input("Dia :"))

print("Aluno 2")
a2 = int(input("Ano :"))
m2 = int(input("Mês :"))
d2 = int(input("Dia :"))

print("mais velho = " , end = "")

if a1 > a2 :
    print("Aluno 2")
elif a2 > a1:
    print("Aluno 1")
elif m1 > m2:
    print("Aluno 2")
elif m2 > m1:
    print("Aluno 1")
elif d1 > d2:
    print("Aluno 2")
elif d2 > d1:
    print("Aluno 1")
else:
    print("mesma idade")

````
