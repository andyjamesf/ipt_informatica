##Mdia dos melhores dois testes

A disciplina de Algoritmos tm trs testes de avaliao.<br>
A nota final  calculada pela mdia dos dois testes com melhor classificao.<br>
Contrua um algoritmo que solicite trs a nota dos testes e de seguida imprima nota final.



```
1 Nota 	: 24.726
2 Nota 	: 31.207
3 Nota 	: 3.889
Nota Final 	= 27.9665
```




````
n1 = float(input("1 nota : "))
n2 = float(input("2 nota : "))
n3 = float(input("3 nota : "))

#n1 e a menor nota
if n1 <= n2 and n1 <= n3:
    #media de n2 e n3
    media = (n2 + n3) / 2
#n2 e a menor nota
elif n2 <= n1 and n2 <= n3:
    media = (n1+n3)/2
else:
    media = (n1 + n2)/2

print("media" , media)



````
