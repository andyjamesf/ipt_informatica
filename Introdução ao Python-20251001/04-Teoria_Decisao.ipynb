## Operadores Relacionais

O resultado de uma expressão que contenha um operador relacional é SEMPRE um valor lógico

Operadores relacionais:


| Símbolo | Significado   |
| ------- | ---------- |
| ==   | igual         |
| !=   | diferente     |
| 	\>	|maior         |
| 	\>=	|maior ou igual|
| 	<	|menor         |
| 	<=	|menor ou igual|


````
n1 =  int(input("N1 = "))
n2 =  int(input("N2 = "))
#cálculo de uma expressão lógica (igualdade == )
igual = n1 == n2;

#impressão de uma variável lógica (igual)
print( n1,  "\t==\t", n2,  "\t->\t", igual )
#cálculo de uma expressão lógica (diferente != )
diferente = n1 != n2;
#impressão de uma variável lógica (diferente)
print(  n1,  "\t!=\t", n2,  "\t->\t", diferente )
#impressão de expressões lógicas
print( n1,  "\t>\t", n2,  "\t->\t", n1 > n2)
print( n1,  "\t>=\t", n2,  "\t->\t", n1 >= n2)
print( n1,  "\t<\t", n2,  "\t->\t", n1 < n2)
print( n1,  "\t<=\t", n2,  "\t->\t", n1 <= n2)

````

## Estrutura de decisão **if**
a instrução if serve para decidir que instruções vão ser executadas baseadas numa expressão lógica
- Se o resultado for verdadeiro (True) executa as instruções do if
- Se o resultado for falso (False) executa as instruções do else
<br>

<pre>
if condição:
    intruções se a condição for verdadeira
else:
    intruções se a condição for falsa
</pre>

Notas:

* a intrução if termina com : <br>
* os blocos de instruções têm de estar indentados <br>
* a intrução else termina com : <br>

#Match case



```
match valor:
  case k1 :
    instruções 1
  case k2 :
    instruções 2
  case ... :
    instruções ...
  case _:
    instruções por defeito

```



````
x = 1
match x :
    case 1:
        print("one")
    case 2:
        print("two")
    case 3:
        print("three")
    case _:
        print("Unknown")
````

##Vários valores no case
o case pode ter vários valores separados pela barra vertical |

````
x = 12
match x :
    case 0|1|2|3|4:
        print("Mau")
    case 5|6|7|8|9:
        print("Mediocre")
    case 10|11|12|13|14:
        print("Médio")
    case 15|16|17|18:
        print("Bom")
    case 19|20:
        print("Muito Bom")
    case _:
        print("Erro")
````
