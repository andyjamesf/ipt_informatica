O python possui instruções que permitem repetir um conjunto de instruções
*   while (Enquanto - repetição condicional)
*   for ( Iteração de um conjunto )
Estas instruções geram ciclos  (loops)


#ciclo While

Executa um conjunto de instruções zero ou mais vezes<br>
O número de vezes que as instruções são executadas deprende da condição.


```
while condição:
    instruções do ciclo
```
Observação
* a condição tem de ter uma variável que é alterada nas instruções
* se a condição for falsa  as instruções do ciclo não são executadas


##Exercício - contar de forma crescente

Algoritmo para escrever os números entre o e 9


![flowchart0[1].png]()

````
#inicializar o contador
i = 0
#condição de paragem
while i < 1000:
    #imprimir o contador
    print(i,  end=" ")
    #atualizar o contador
    i= i + 1
#instrução fora do ciclo
print("\nacabei")
````

## Dígitos de um numero

Construa um algoritmo para imprimir na consola os dígitos de um numero inteiro.

![folder[1].jpg]()  ![flowchart0[1].png]()


````
num = int(input("Número : "))
print("Dígitos = " , end  ="")
while num > 0 :
    #imprimir o último digito
    print( num % 10 , end = " ")
    #dividir o numero por 10 ( // divisão inteira)
    num //= 10

````

# Converter Decimal para Binário

Escreva um algoritmo para converter para binario um numero introduzido pelo utilizador.



![folder[1].jpg]() ![flowchart0[1].png]()


````
#ler o numero decimal
num = int(input("Decimal : "))
#inicializar o numero binario
binario = ""
#enquanto houver digitos
while num > 0:
    #calcular o bit (% resto da divisão)
    bit = num %2
    #adicionar o digito no inicio do binario
    binario = str(bit) + binario
    #retirar o bit do numero (//divisao inteira)
    num //= 2

print("Binario = " , binario)


````

## Exercício **Divisores de um numero**
Construa um algoritmo que imprima na consola os divisores de um numero.
O numero deve ser introduzido pelo utilizador.

![folder[1].jpg]()


````
n = int(input("Número : "))
i = 1
print("Divisores : ", end = "")
while i <= n :
    if n%i == 0:
        print(i , end = " ")

    i = i+1
````
